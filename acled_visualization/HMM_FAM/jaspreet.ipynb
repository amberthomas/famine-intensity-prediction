```python
frame = pd.read_csv('jaspreet_afghanistan.csv')

print(len(frame.Country))

year1 = (frame.jaspreet_2013)
year2 = (frame.jaspreet_2014)
year3 = (frame.jaspreet_2015)
year4 = (frame.jaspreet_2016)
year5 = (frame.jaspreet_2017)



yeardata = np.empty([1,1])


yeardata = np.vstack((year1,year2,year3,year4,year5))


yeardata = yeardata.T - 1
#print(yeardata)
yeardata = yeardata[yeardata.min(axis=1)>=0,:]

print("new shape is")
print(yeardata.shape)
yeardata = yeardata.tolist()
print(yeardata)
print("final shape is")
print(len(yeardata))
#newyear = yeardata
#yeardata = yeardata[:,0:3]
#print(yeardata)

model = HiddenMarkovModel.from_samples(DiscreteDistribution, n_components=5, X=yeardata ,algorithm='baum-welch')
#model.predict(yeardata, algorithm='viterbi')
#model = model.fit(yeardata[1:33,0:3], algorithm='viterbi')

count = 0
output_diagnol = []
output = {}
#of the type actual : predicted
temp = 1
for i in range(temp,len(yeardata)):
    print("yeardata is")
    print(yeardata[i])
    cool = model.predict_proba(yeardata[i,:])
    #print("cool is")
    #print(cool)
    cool1 = model.predict(yeardata[i,:], algorithm='map')
    cool2 = model.viterbi(yeardata[i,:])
    print("cool2 is")
    print(cool2)
    #print("shape of cool is")
    #print(cool.shape)
    if(cool1[1] == yeardata[i,4]):
        output_diagnol.append(yeardata[i,4])
        print("Theres a match!")
        #print("yeardata is")
        #print(yeardata[i,4])
        #key = yeardata[i,4]
        #output[key] = np.argmax(cool[4,:],axis =0)
        count+=1
    else:
        #print("else part here")
        #print(i)
        key = yeardata[i,4]
        #print("key is")
        #print(key)
        #print("value is")
        #print(np.argmax(cool[4,:],axis =0))
        output[key] = (cool1[1])                       
        
print("correct count is")
print(count)
print("accuracy for Afghanistan is") 
print(count/(len(frame.Country)-temp))
print(output)
print(output_diagnol)

        
#temp = 21, accuracy for Afghanistan is
#0.46153846153846156
#temp=26, -> 0.5
#temp=31 -> 0.666

#5 -> 9.09
#2 -> 36.36
#3 -> 42.42
```

```python
frame = pd.read_csv('jaspreet_niger.csv')

print(len(frame.Country))

year1 = (frame.jaspreet_2014)
year2 = (frame.jaspreet_2015)
year3 = (frame.jaspreet_2015_2)
year4 = (frame.jaspreet_2016)
year5 = (frame.jaspreet_2016_2)
year6 = (frame.jaspreet_2017)
year7 = (frame.jaspreet_2017_2)
year8 = (frame.jaspreet_2018)



yeardata = np.empty([1,1])


yeardata = np.vstack((year1,year2,year3,year4,year5,year6,year7,year8))


yeardata = yeardata.T - 1
#print(yeardata)

yeardata = yeardata[yeardata.min(axis=1)>=0,:]

print("new shape is")
print(yeardata.shape)
#print(yeardata[1:30,:])
model = HiddenMarkovModel.from_samples(DiscreteDistribution, n_components=2, X=yeardata[1:47,0:6])
#model.predict(yeardata, algorithm='viterbi')
#output1 = model.fit(yeardata[1:47,0:6], algorithm='baum-welch')
#print(yeardata[7,:])
count = 0
output_diagnol = []
output = {}
#of the type actual : predicted
temp = 1
for i in range(temp,yeardata.shape[0]):
    cool = model.predict_proba(yeardata[i,:])
    #print("cool shape")
    #print(cool.shape)
    
    if(np.argmax(cool[7,:],axis =0) == yeardata[i,7]):
        output_diagnol.append(yeardata[i,7])
        
        count+=1
    else:
        key = yeardata[i,7]
        output[key] = (np.argmax(cool[7,:],axis =0))                       
        
        
print("accuracy for Niger is") 
print(count/(yeardata.shape[0]-temp))
print(output)
print(output_diagnol)
#temp = 41
#accuracy = 5
#temp=46
#accuracy=6.66
#temp=51
#acc = 10

####after removing zeroes
#31 - 5.88
#26- 9.09

#5 compos 

#2 compos = 59.57
#5 = 17.02
#3 -> 40.42
```
