```python
frame = pd.read_csv('jaspreet_afghanistan.csv')

print(len(frame.Country))

year1 = (frame.jaspreet_2013)
year2 = (frame.jaspreet_2014)
year3 = (frame.jaspreet_2015)
year4 = (frame.jaspreet_2016)
year5 = (frame.jaspreet_2017)



yeardata = np.empty([1,1])

#print(yeardata)
yeardata = np.vstack((year1,year2,year3,year4,year5))
#yeardata.append(year2)
#yeardata.append(year3)
#yeardata.append(year4)
#yeardata.append(year5)
#print(yeardata.T)

yeardata = yeardata.T - 1
print(yeardata[33,:])
#print(yeardata[1:30,:])
model = HiddenMarkovModel.from_samples(DiscreteDistribution, n_components=5, X=yeardata[1:30,:])
#model.predict(yeardata, algorithm='viterbi')
output = model.fit(yeardata[1:30,:], algorithm='viterbi')
#print(yeardata[7,:])

#print("cool is")
#print(cool)
#print(cool[4,:])
count = 0
output_diagnol = []
output = {}
#of the type actual : predicted
temp = 31
for i in range(temp,len(frame.Country)):
    cool = model.predict_proba(yeardata[i,:])
    print(i)
    #print("cool is")
    #print(cool)
    if(np.argmax(cool[4,:],axis =0) == yeardata[i,4]):
        output_diagnol.append(yeardata[i,4])
        #key = yeardata[i,4]
        #output[key] = np.argmax(cool[4,:],axis =0)
        count+=1
    else:
        print("else part here")
        print(i)
        key = yeardata[i,4]
        #print("key is")
        #print(key)
        #print("value is")
        #print(np.argmax(cool[4,:],axis =0))
        output[key] = (np.argmax(cool[4,:],axis =0))                       
        
        
print("accuracy for Afghanistan is") 
print(count/(len(frame.Country)-temp))
print(output)
print(output_diagnol)

        
#temp = 21, accuracy for Afghanistan is
#0.46153846153846156
#temp=26, -> 0.5
#temp=31 -> 0.666
```
